```python
import cv2
import matplotlib.pyplot as plt

from diploma.models.detector import FaceDetectorONNX
from diploma.drawing import draw_landmarks, draw_rectangles
```

```python
detector = FaceDetectorONNX("E:\kbtu_courses\diploma_project\src\diploma\models\detector\weights\model_detector.onnx")

image = cv2.imread("E:\kbtu_courses\diploma_project\src\diploma\photos\SCRFD_check\Big_photo.jpg")
boxes, landmarks = detector(image, input_size=(640, 640))
```

```python
print(boxes[0])
print(landmarks[0])
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(draw_rectangles(image, bboxes = boxes, override=True))
plt.imshow(draw_landmarks(image, landmarks=landmarks))
```
